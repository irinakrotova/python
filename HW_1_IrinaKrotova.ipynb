### Задача 2 (2 балла)

Посчитать 1! + 2! + 3! + 4! + ... + n! для заданного n. Решить только одним циклом

```python
def function_2(n):
    #Реализуйте последовательность действий, которая будет преобразовать число n в result
    result=0
    a=1
    for k in range (1,n+1):
        a*=k
        result+=a
    return result
    #Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
assert function_2(1) == 1
assert function_2(3) == 9
assert function_2(0) == 0
assert function_2(-1) == 0
```

### Задача 3 (2 балла)

Найти среднее арифметическое отрицательных элементов в списке

```python
#list_of_numbers - список чисел
#mean - среднее арифметическое отрицательных элементов
def function_3(list_of_numbers):
    s, l = 0, 0
    for i in list_of_numbers:
        if i<0:
            s += i
            l += 1
    if s == 0:
        mean=0
    else :
        mean=s/l
    
    #тут реализуйте вычесление mean по list_of_numbers
    return mean
#Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
assert function_3([]) == 0
assert function_3([1, -1, -3, 3]) == -2
assert function_3([1, 2, 3]) == 0
# т.д. например, что будет, если список не содержит отрицательных элементов
```

### Задача 4 (2 балла)

Найти расстояние между точками с заданными координатами в n-мерном пространстве

```python
def function_4(x, y):
    s=0
    
    #тут реализуйте вычисление расстояния distance между точками point_1 и point_2
    for i,j in zip(x,y):
       s+=(i-j)**2
    
    return s**0.5
#Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
assert function_4((1, 0), (2, 0)) == 1
assert function_4((1, 7, 5),(4, 7, 9)) == 5
```

```python
def function_4(binary_list):
    #место для вашей реализации
    begin=-1
    end_index=-2
    begin_index=-1
    for i,n in enumerate(binary_list+[0]):
        if n == 0:
            if begin>=0 and i-begin > end_index-begin_index+1:
                begin_index=begin
                end_index=i-1
            begin=-1             
        else:
            if begin == -1:
                begin=i
    len_ones_sequence=end_index - begin_index+1
    if begin_index != -1:
        return len_ones_sequence, begin_index, end_index

#Тесты
assert function_4([0, 1, 0, 1, 1, 0, 0, 0]) == (2, 3, 4)
assert function_4([0, 0, 0, 0, 0, 0, 0, 0]) == None
assert function_4([1, 1, 1, 1, 0, 1, 1, 1]) == (4, 0, 3)
assert function_4([]) == None
assert function_4([0, 1, 0, 1, 1, 1, 1, 1]) == (5, 3, 7)
```

### Задача 7 (3 балла)

На вход подаётся строка S и буква w. Нужно развернуть в этой строке подстроку, заключенную между первой и последней буквой w, и вернуть получившуюся строку. Если данной буквы в строке нет, то вернуть -1. Если строка содержит только одну данную букву, то вернуть строку без изменений

```python
def function_7(S, w):
    s=S.split(w)
    join=w.join(s[1:-1])
    join=list(join)
    join.reverse()
    join=''.join(join)
    new_S=s[0]+w+join+w+s[-1]
    return new_S

#Тесты
assert function_7('abcdeeefgefg', 'e') == 'abcdegfeeefg'
assert function_7('abcdefgefg', 'e') == 'abcdegfefg'
assert function_7('abZdcfgcfg', 'c') == 'abZdcgfcfg'
assert function_7('cabcdcfgcfgc', 'c') == 'cgfcgfcdcbac'
```

### Задача 8 (4 балла + 2 балла за оформления в виде функции, принимающей текст на русском языке)

Дан текст. Посчитать, сколько раз встречается каждое словои и вывести слова в порядке убывания количествА повторений слова (при совпадении частоты встречаемости -- вывести в лексикографическом порядке)

```python
def count_words (text):
    dictionary = {}
    for word in text.split():
        word1=''
        word=word.lower()
        for i in word:
            if i.isalpha() or i == '-':
                word1+=i
        if word1 != '':
            dictionary.setdefault(word1, 0)
            dictionary[word1] += 1
    sort_s=sorted([(z,y) for y,z in dictionary.items()])
    return(sort_s)

with open('problem_8_text.txt', 'r') as fl:
     text = fl.read()

for g,h in count_words(text):
    print('%20s || %d' %(h, g))
    
#Теперь в text лежит сказка о Бабе-Яге =)

#Теперь обработайте текст и сделайте красивенький вывод
```
